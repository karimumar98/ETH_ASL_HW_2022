```python
class cache:
    def __init__(self):
        self.tags = [-1 for _ in range(8)]
        self.contents = [None for _ in range(8)]
        
    def print_contents(self):
        for i in range(8):
            print(self.contents[i])
    def latex_table(self):
        for i in range(8):
            if self.contents[i] == None:
                print(f"{i} & $ Empty $ & $ Empty $&  \\\\")
            else:
                print(f"{i} & $ {self.contents[i][0]} $ & $ {self.contents[i][1]}$&  \\\\")
    
    def get(self, addr, i, j, var):
        #print (f"getting data at address {addr}.. tag: {self.get_tag(addr)}, set: {self.get_set(addr)}, block offset: {self.get_block_offset(addr)}")
        if self.tags[self.get_set(addr)] == self.get_tag(addr):
            return True
        else:
            self.tags[self.get_set(addr)] = self.get_tag(addr)
            self.contents[self.get_set(addr)] = (mem[addr- self.get_block_offset(addr)], mem[addr- self.get_block_offset(addr) + 8])
            return False
        
    def get_tag(self, addr):
        return addr >> 7
    
    def get_block_offset(self, addr):
        return addr%16
    
    def get_set (self, addr):
        return (addr%128) >> 4 
```

```python
class cacheb:
    
    # 4 sets
    # 2 way set associative
    def __init__(self):
        self.tags = [[-1, -1] for _ in range(8)]
        self.contents = [[None, None] for _ in range(8)]
        self.lru = [-1 for _ in range(4)] # which line was least recently used
    def print_contents(self):
        for i in range(4):
            print(self.contents[i])
            
    def latex_table(self):
        for i in range(4):
            if self.contents[i][0] == None:
                print(f"{i} & $ Empty $ & $ Empty $&  \\\\")
            elif self.contents[i][1] == None:
                print(f"{i} & $ {self.contents[i][0][0]}, {self.contents[i][0][1]} $ & $ Empty$&  \\\\")
            else:
                print(f"{i} & $ {self.contents[i][0][0]}, {self.contents[i][0][1]} $ & $ {self.contents[i][1][0]}, {self.contents[i][1][1]}$&  \\\\")
        
    def get(self, addr, i, j, var):
        #print (f"-----getting data at address {addr}.. tag: {self.get_tag(addr)}, set: {self.get_set(addr)}, block offset: {self.get_block_offset(addr)}")
        #print("current state:")
        #self.print_contents()
        if self.lru[self.get_set(addr)] == -1:
            ## Case cache empty -> put data into first and set LRU = 1
            self.lru[self.get_set(addr)] = 1
            self.tags[self.get_set(addr)][0] = self.get_tag(addr)
            self.contents[self.get_set(addr)][0] = (mem[addr], mem[addr + 8])
            
        elif self.tags[self.get_set(addr)][0] == self.get_tag(addr) or self.tags[self.get_set(addr)][1] == self.get_tag(addr):
            ## Cache line in one of the sets
            return True
        else:
            # We have to evict a line
            print(f"Evicting {self.contents[self.get_set(addr)][self.lru[self.get_set(addr)]]}")
            self.tags[self.get_set(addr)][self.lru[self.get_set(addr)]] = self.get_tag(addr)
            self.contents[self.get_set(addr)][self.lru[self.get_set(addr)]] = (mem[addr- self.get_block_offset(addr)], mem[addr- self.get_block_offset(addr) + 8])
            
            ## Change last used cache
            self.lru[self.get_set(addr)] = 1-self.lru[self.get_set(addr)]
            return False
        
    def get_tag(self, addr):
        return addr >> 6
    
    def get_block_offset(self, addr):
        return addr%16
    
    def get_set (self, addr):
        return (addr%64) >> 4
```
